# Preview notebook only

This notebook is for inspecting already-generated nominal-rigidity outputs and trying temporary table previews. The production source is `main.py`, run through `code/Makefile`. Do not use this notebook to regenerate paper artifacts.

# Nominal Rigidities

Run from the task code directory or from the repository root. Edit the play-area variables, then rerun the cell to inspect alternative samples, outcomes, or regressors.

```python
from pathlib import Path
import sys
from io import StringIO
from tempfile import TemporaryDirectory

import pandas as pd
from IPython.display import display

TASK_NAME = "nominal_rigidities"
SPEC_KEY = "nominal_rigidities"

code_dir = Path.cwd()
if code_dir.name != "code":
    code_dir = Path("tasks/measurement/nominal_rigidities/code").resolve()
task_dir = code_dir.parent
sys.path.append(str(task_dir.parent))

from regression_table_tools import build_table, load_regression_data, load_table_params

params = load_table_params(code_dir / "main.py")
spec = params["table_specs"][SPEC_KEY]
df = load_regression_data(task_dir / "input", params)

# Play area: change these and rerun the cell.
samples_to_run = ["full_sample", "estimated_models"]
outcomes_to_run = list(params["outcomes"])
variables = list(spec["variables"])

preview_params = dict(params)
preview_params["outcomes"] = outcomes_to_run
samples = {
    "full_sample": df,
    "estimated_models": df.loc[df["estimated"] == 1].copy(),
}

print(f"Rows after timing restrictions: {len(df)}")
print(f"Rows in estimated-model sample: {len(samples['estimated_models'])}")
print("Variables in this exercise:")
print(variables if variables else "No additional variables; fixed controls are added from config.")
print("Outcomes:")
print(outcomes_to_run)

for sample_name in samples_to_run:
    sample_df = samples[sample_name]
    sample_label = sample_name.replace("_", " ")
    title = f"{spec['title']} ({sample_label})"

    with TemporaryDirectory() as tmpdir:
        table_path = Path(tmpdir) / "preview_table.txt"
        build_table(sample_df, variables, title, table_path, preview_params)
        table_text = table_path.read_text(encoding="utf-8")

    print("\n" + title)
    print("-" * len(title))
    lines = table_text.splitlines()
    header_i = next(i for i, line in enumerate(lines) if line.startswith("Variable\t"))
    table_body = "\n".join(lines[header_i:])
    table_df = pd.read_csv(StringIO(table_body), sep="\t").fillna("")
    display(table_df)
    print(table_text)
```
