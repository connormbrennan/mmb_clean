# LASSO Feature Screen

This notebook is for inspecting the LASSO feature screens and the printable post-selection tables. The production build lives in `build_lasso.py`; edit the constants near the top of that script to change the candidate covariates, sparse-feature rule, or penalty grids.


```python
from pathlib import Path

import pandas as pd
from IPython.display import display

CODE_DIR = Path.cwd() if Path.cwd().name == "code" else Path("tasks/measurement/lasso/code").resolve()
TASK_DIR = CODE_DIR.parent
OUTPUT_DIR = TASK_DIR / "output" / "lasso"

print(OUTPUT_DIR)
```

## Optional Rebuild

Set `RUN_BUILD = True` to rerun the full unweighted and citation-weighted LASSO screens from inside the notebook. It can take a minute because the Huber LASSO uses cross-validation.


```python
RUN_BUILD = False

if RUN_BUILD:
    %run build_lasso.py
```

## Selection Summaries


```python
summary = pd.read_csv(OUTPUT_DIR / "lasso_zeroed_summary.csv")
summary_cw = pd.read_csv(OUTPUT_DIR / "lasso_zeroed_summary_citation_weighted.csv")

print("Unweighted")
display(summary)
print("Citation weighted")
display(summary_cw)
```

## Selected Features


```python
nonzero = pd.read_csv(OUTPUT_DIR / "lasso_nonzero_features.csv")
nonzero_cw = pd.read_csv(OUTPUT_DIR / "lasso_nonzero_features_citation_weighted.csv")

print("Unweighted selected features")
display(nonzero.sort_values(["outcome", "abs_coefficient"], ascending=[True, False]))
print("Citation-weighted selected features")
display(nonzero_cw.sort_values(["outcome", "abs_coefficient"], ascending=[True, False]))
```

## Post-Selection Regression Tables

These tables refit the LASSO-selected variables with policy-rule and estimated-model controls. Standard errors are clustered by model.


```python
post = pd.read_csv(OUTPUT_DIR / "lasso_post_selection_table.csv")
post_cw = pd.read_csv(OUTPUT_DIR / "lasso_post_selection_table_citation_weighted.csv")

print("Unweighted post-selection source rows")
display(post.loc[post["selected"]].sort_values(["outcome", "abs_screen_coefficient"], ascending=[True, False]))
print("Citation-weighted post-selection source rows")
display(post_cw.loc[post_cw["selected"]].sort_values(["outcome", "abs_screen_coefficient"], ascending=[True, False]))
```

## Printable LaTeX


```python
for filename in [
    "lasso_post_selection_table.tex",
    "lasso_post_selection_table_citation_weighted.tex",
]:
    print("\n" + "=" * 80)
    print(filename)
    print("=" * 80)
    print((OUTPUT_DIR / filename).read_text(encoding="utf-8"))
```
